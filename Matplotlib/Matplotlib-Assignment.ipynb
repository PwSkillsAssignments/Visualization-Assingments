```python
import numpy as np

np.random.seed(3)

x = 3 + np.random.normal(0, 2, 50)

y = 3 + np.random.normal(0, 2, len(x))
```

```python
plt.figure(figsize=(15,10))
plt.scatter(x=x, y=y)
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.show()
```

Q4: What is a bar plot? Why is it used? Using the following data plot a bar plot and a horizontal bar plot. 

import numpy as np

company = np.array(["Apple", "Microsoft", "Google", "AMD"])

profit = np.array([3000, 8000, 1000, 10000])


```python
company = np.array(["Apple", "Microsoft", "Google", "AMD"])

profit = np.array([3000, 8000, 1000, 10000])
```

```python
plt.barh(y=company, width=profit, color='#805F0F')
plt.xlabel('Profit')
plt.ylabel('Companies')
plt.title('Profits of the Company')
plt.show()
```

Q5: What is a box plot? Why is it used? Using the following data plot a box plot.

box1 = np.random.normal(100, 10, 200)

box2 = np.random.normal(90, 20, 200)

```python
box1 = np.random.normal(100, 10, 200)

box2 = np.random.normal(90, 20, 200)
```

```python
# boxplot represents the distribution of the dataset, it summarizes  the min, Q1, median, Q3 and max.
# It is used to spot the outliers, identify Central tendency

# Creating a box plot
data = [box1, box2]
labels = ['Box 1', 'Box 2']

plt.boxplot(data, labels=labels)

# Adding labels and title
plt.xlabel('Box')
plt.ylabel('Values')
plt.title('Box Plot of Box1 and Box2')

# Display the plot
plt.show()
```
